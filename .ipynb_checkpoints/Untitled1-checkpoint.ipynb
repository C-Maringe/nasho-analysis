```python
import numpy as np
import pandas as pd

# Generate dataset with 6 features and 1 target variable
X, y = make_regression(n_samples=67, n_features=6, n_informative=6, n_targets=1,
                       noise=0.0, random_state=42)

# Scale the features to be in the range of 1 to 5
X_scaled = (X - np.min(X, axis=0)) / (np.max(X, axis=0) - np.min(X, axis=0)) * 4 + 1

# Scale the target variable to be in the range of 0 to 15%
y_scaled = y / np.max(y) * 0.15

# Convert NumPy arrays to Pandas DataFrame
df = pd.DataFrame(np.column_stack((X_scaled[:,0], X_scaled[:,1], X_scaled[:,2], 
                                    X_scaled[:,3], X_scaled[:,4], X_scaled[:,5], y_scaled)),
                  columns=['asset_base', 'bda_capability', 'predictive_analytics_usage', 
                           'descriptive_analytics_usage', 'diagnostic_analytics_usage', 
                           'prescriptive_analytics_usage', 'revenue_growth'])

# View the first few rows of the DataFrame
df.head()
```

```python
X = df[['asset_base', 'bda_capability', 'predictive_analytics_usage', 
                           'descriptive_analytics_usage', 'diagnostic_analytics_usage', 
                           'prescriptive_analytics_usage']]
y = df['revenue_growth']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
r_squared = model.rsquared
print(r_squared)
```
